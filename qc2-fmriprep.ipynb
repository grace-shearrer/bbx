#### Setup a dictionary to hold fmriprep qc values  
Currently we are counting the available training and resting functionals found.

```python
fmriprep_dict = {}
for sub_id in sorted(bids_subs):

    if sub_id not in fmriprep_dict:
        fmriprep_dict[sub_id] = {}
        
    func_run1 = glob.glob(os.path.join(bids_path, "derivatives/fmriprep", sub_id, "ses-1/func/*training*run-1*-preproc_bold.nii.gz"))
    func_run2 = glob.glob(os.path.join(bids_path, "derivatives/fmriprep", sub_id, "ses-1/func/*training*run-2*-preproc_bold.nii.gz"))
    func_run3 = glob.glob(os.path.join(bids_path, "derivatives/fmriprep", sub_id, "ses-1/func/*training*run-3*-preproc_bold.nii.gz"))
    func_run4 = glob.glob(os.path.join(bids_path, "derivatives/fmriprep", sub_id, "ses-1/func/*training*run-4*-preproc_bold.nii.gz"))
    funcs_resting = glob.glob(os.path.join(bids_path, "derivatives/fmriprep", sub_id, "ses-1/func/*resting*-preproc_bold.nii.gz"))
    anat_files = glob.glob(os.path.join(bids_path, "derivatives/fmriprep", sub_id, "anat/{}_space-MNI152NLin2009cAsym_desc-preproc_T1w.nii.gz".format(sub_id)))
        
    fmriprep_dict[sub_id]["run1_func"] = len(func_run1)
    fmriprep_dict[sub_id]["run2_func"] = len(func_run2)
    fmriprep_dict[sub_id]["run3_func"] = len(func_run3)
    fmriprep_dict[sub_id]["run4_func"] = len(func_run4)
    fmriprep_dict[sub_id]["resting_func"] = len(funcs_resting)
    fmriprep_dict[sub_id]["total_funcs"] = len(funcs_resting) + len(func_run1) + len(func_run2) + len(func_run3) + len(func_run4)
    fmriprep_dict[sub_id]["anat_niftis"] = len(anat_files)
```

```python
bids_path = '/projects/niblab/bids_projects/Experiments/bbx/bids'
bad_fd_subs=[]
missing_regressors=[]

# FD check
def write_report(sub_id, run_id, comparator, fd_total_ct, nifti, bad_subs):
    bad_fd_subs.append(sub_id)  
    line="\nFailed FD check: \nsub: {} \trun: {}  \tcomparator: {} \tfd_count: {} \nNifti: {}\n".format(sub_id,
                                         run_id, comparator, fd_total_ct, nifti.split("/")[-1])
    print(line)
    fd_file.write(line)
    #return bad_fd_subs

sub_ids = [x.split("/")[-2] for x in glob.glob(os.path.join(bids_path, 'derivatives/fmriprep/sub-*/ses-1'))]
for sub_id in sorted(sub_ids):
    tsv_files=glob.glob(os.path.join(bids_path, 'derivatives/fmriprep/{}/ses-1/func/sub-*_ses-1_task-training_run-*_desc-confounds_regressors.tsv'.format(sub_id)))
    
    for tsv in sorted(tsv_files):
        
        # set initial variables
        fd_total_ct = 0
        sub_id=tsv.split("/")[-1].split("_")[0]
        #print(tsv)
        run_id=tsv.split("_")[4]
        nifti=os.path.join(bids_path,
                           'derivatives/fmriprep/{}/ses-1/func/{}_ses-1_task-training_{}_space-MNI152NLin2009cAsym_desc-preproc_bold.nii.gz '.format(sub_id, sub_id, run_id))
        shell_cmd1 = 'fslnvols {}'.format(nifti) 
        volume = sp.check_output(shell_cmd1, shell=True, encoding="utf-8")
        volume = volume.strip()
        comparator = int(volume) *.25
        
        # load dataframe
        temp_df = pd.read_csv(tsv, sep="\t")
        
        body_motcor_params = temp_df[['trans_x', 'trans_y', 'trans_z', 'rot_x', 'rot_y', 'rot_z' ]]
        #print(body_motcor_params)
        
        try:
            
            spike_regressors = temp_df[['motion_outlier00','motion_outlier01', 'motion_outlier02', 'motion_outlier03','motion_outlier04', 'motion_outlier05']]
            #print(spike_regressors)
        except:
            # add report here
            #line="\n\n> Spike Regressors not found for {} | {} \n".format(sub_id, run_id)
            #fd_file.write(line)
            print("\n\n> Spike Regressors not found for " + sub_id +" " +run_id)
            missing_regressors.append((sub_id, run_id))

        
        # analyze FD values
        max_fd = temp_df['framewise_displacement'].max()
        min_fd = temp_df['framewise_displacement'].min()
        
    
        for fd_val in  temp_df['framewise_displacement']:
            if fd_val > 0.9:
                #print("bad fd value, ", fd_val)
                fd_total_ct += 1
        
        # fill in dictionary       
        if run_id == "run-1":
            scan_df.loc[sub_id, "min_r1"]=min_fd
            scan_df.loc[sub_id, "max_r1"]=max_fd
            
            if max_fd < 3.0:
                scan_df.loc[sub_id, "mcflirt_3mm_r1"]="pass"
            else:
                scan_df.loc[sub_id, "mcflirt_3mm_r1"]="fail"
                
            if fd_total_ct < comparator:
                scan_df.loc[sub_id, "fd_r1"]="pass"
            else:
                scan_df.loc[sub_id, "fd_r1"]="fail"
                write_report(sub_id, run_id, comparator, fd_total_ct, nifti, bad_fd_subs)
                
        elif run_id == "run-2":
            scan_df.loc[sub_id, "min_r2"]=min_fd
            scan_df.loc[sub_id, "max_r2"]=max_fd
            
            if max_fd < 3.0:
                scan_df.loc[sub_id, "mcflirt_3mm_r2"]="pass"
            else:
                scan_df.loc[sub_id, "mcflirt_3mm_r2"]="fail"
                
            if fd_total_ct < comparator:
                scan_df.loc[sub_id, "fd_r2"]="pass"
            else:
                scan_df.loc[sub_id, "fd_r2"]="fail"
                write_report(sub_id, run_id, comparator, fd_total_ct, nifti, bad_fd_subs)
                
        elif run_id == "run-3":
            scan_df.loc[sub_id, "min_r3"]=min_fd
            scan_df.loc[sub_id, "max_r3"]=max_fd
            
            if max_fd < 3.0:
                scan_df.loc[sub_id, "mcflirt_3mm_r3"]="pass"
            else:
                scan_df.loc[sub_id, "mcflirt_3mm_r3"]="fail"
                
                
            if fd_total_ct < comparator:
                scan_df.loc[sub_id, "fd_r3"]="pass"
            else:
                scan_df.loc[sub_id, "fd_r3"]="fail"
                write_report(sub_id, run_id, comparator, fd_total_ct, nifti, bad_fd_subs)
                
        else:
            scan_df.loc[sub_id, "min_r4"]=min_fd
            scan_df.loc[sub_id, "max_r4"]=max_fd
            
            if max_fd < 3.0:
                scan_df.loc[sub_id, "mcflirt_3mm_r4"]="pass"
            else:
                scan_df.loc[sub_id, "mcflirt_3mm_r4"]="fail"
                
                
            if fd_total_ct < comparator:
                scan_df.loc[sub_id, "fd_r4"]="pass"
            else:
                scan_df.loc[sub_id, "fd_r4"]="fail"
                write_report(sub_id, run_id, comparator, fd_total_ct, nifti, bad_fd_subs)
            

```
